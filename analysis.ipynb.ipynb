```python
#pandas operations
data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [80, 90, 85, 70]
}

df = pd.DataFrame(data)

print(df)
```

```python
#inspecting data
#view first few rows
print(df.head())
```

```python
#summary of the data
print(df.describe())
```

```python
#filtering and sorting
filtered_df = df[df["Marks"] > 85]
print(filtered_df)
```

```python
#sort by age
sorted_df = df.sort_values(by="Marks")
print(sorted_df)
```

```python
#data Manipulation
#add new column
df["Result"] = df["Marks"] > 75
print(df)
```

```python
#Update Values
df["Marks"] = df["Marks"] + 5
print(df)
```

```python
#Filter Data
print(df[df["Marks"] > 80])
```

```python
#sorting
print(df.sort_values(by="Marks", ascending=False))
```

```python
#add grade column
def grade(m):
    if m >= 90:
        return "A"
    elif m >= 80:
        return "B"
    else:
        return "C"

df["Grade"] = df["Marks"].apply(grade)
print(df)
```

```python
#drop column
df = df.drop("Result", axis=1)
print(df)
```

```python
#rename column
df.rename(columns={"Marks": "Score"}, inplace=True)
print(df)
```

```python
#Handling Missing Values
df.isnull()
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [80, np.nan, 85, None]
}

df = pd.DataFrame(data)
print(df)
```

```python
#integrating numpy with pandas
import pandas as pd
import numpy as np

data = {
    "Marks": [80, 90, 85, 70, 95]
}

df = pd.DataFrame(data)
```

```python
mean = np.mean(df["Marks"])
median = np.median(df["Marks"])
sd = np.std(df["Marks"])

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", sd)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"Marks": [80, 90, 85, 70, 95]})

mean = np.mean(df["Marks"])
std = np.std(df["Marks"])

df["Z_score"] = (df["Marks"] - mean) / std

print(df)
```

```python
#loading Data:
import pandas as pd

df = pd.read_csv("data.csv")
df.head()
```

```python
#cleaning the data
df = df.dropna()
df = df.drop_duplicates()
print(df)
```

```python
# Aggregate
avg_salary = df.groupby("Department")["Salary"].mean()
print(avg_salary)
```

```python
# Visualize
import matplotlib.pyplot as plt

df["Salary"].plot()
plt.show()
```

```python
#Bar Chart
df.groupby("Department")["Salary"].mean().plot(kind="bar")
plt.show()
```

```python
#Histogram
df["Salary"].plot(kind="hist")
plt.show()
```
